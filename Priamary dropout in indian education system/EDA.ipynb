```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
subset_drop.describe()
```

```python
subset_drop["UP_Boys"]=  subset_drop["UP_Boys"].replace(0.00,4.29)
subset_drop["Primary_Boys"] = subset_drop["Primary_Boys"].replace(0.00,4.29)
subset_drop["Primary_Girls"] = subset_drop["Primary_Girls"].replace(0.00,4.01)
subset_drop["Primary_Total"] = subset_drop["Primary_Total"].replace(0.00,4.15)
subset_drop["Upper Primary_Girls"] = subset_drop["Upper Primary_Girls"].replace(0.00,4.16)
subset_drop["Upper Primary_Total"] = subset_drop["Upper Primary_Total"].replace(0.00,3.83)
subset_drop.head(17)
```
